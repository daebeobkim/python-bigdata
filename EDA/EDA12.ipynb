```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
```

```python
d1 = np.loadtxt("C:/Users/rlaeo/Downloads/1D/example_1.txt")
d2 = np.loadtxt("C:/Users/rlaeo/Downloads/1D/example_2.txt")
print(d1.shape,d2.shape)
```

Histogram Plots

```python
plt.hist(d1,label="D1")
plt.hist(d2,label="D2")
plt.legend()
plt.ylabel("counts")
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)#순서대로 구간시작점, 구간끝점, 구간 내 숫자 갯수
counts1,_,_ = plt.hist(d1,bins=bins,label="D1", density = True)
plt.hist(d2, bins=bins, label="D2")
plt.legend()
plt.ylabel("Counts") #확률상으로 말할 수 있는데 10에서 15사이에서 d1이 나올 확률이 더 높고, 10에서 15사이에서 d2가 나올 확률이 높다.
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)#순서대로 구간시작점, 구간끝점, 구간 내 숫자 갯수
counts1,_,_ = plt.hist(d1,bins=bins,label="D1", density = True)# density = True 사용시 %로 출력
plt.hist(d2, bins=bins, label="D2",density = True)
plt.legend()
plt.ylabel("Probability")
plt.show()
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)#순서대로 구간시작점, 구간끝점, 구간 내 숫자 갯수
counts1,_,_ = plt.hist(d1,bins=bins,label="D1", density = True,histtype = "step", lw = 3)# density = True 사용시 %로 출력
plt.hist(d2, bins=bins, label="D2",density = True,histtype = "step",ls=":")
plt.legend()
plt.ylabel("Probability")
plt.show()
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)#순서대로 구간시작점, 구간끝점, 구간 내 숫자 갯수
plt.hist([d1,d2], bins=bins, label="Stacked",density = True,alpha=0.5)
plt.hist(d1,bins=bins,label="D1", density = True,histtype = "step", lw = 3)# density = True 사용시 %로 출력
plt.hist(d2, bins=bins, label="D2",density = True,histtype = "step",ls=":")
plt.legend()
plt.ylabel("Probability")
plt.show()
```
