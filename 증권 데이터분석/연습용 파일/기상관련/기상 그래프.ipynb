```python
import csv
import matplotlib.pyplot as plt
f = open('C:/Users/rlaeo/OneDrive/바탕 화면/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
next (data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '': #최고 기온 값과 최저 기온 값이 존재한다면
        data = row[0].split('-')    #날짜 값을 - 문자를 기준으로 구분하여 저장
        if 2000 <= int(data[0]):    #2000년 이후 데이터라면
            if data[1] == '04' and data[2] == '14':     #4월 14일 이라면
                high.append(float(row[-1]))     #최고기온 high 리스트에 저장
                low.append(float(row[-2]))      #최저 기온 low 리스트에 저장

plt.rc('font',family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False #-기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프') #제목 설정

plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.legend() #범례표시
plt.show()
```
