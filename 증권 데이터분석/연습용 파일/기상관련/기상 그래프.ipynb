```python
import csv
import matplotlib.pyplot as plt
f = open('C:/Users/rlaeo/OneDrive/ /seoul.csv','r',encoding='cp949')
data = csv.reader(f)
next (data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '': #      
        data = row[0].split('-')    #  -    
        if 2000 <= int(data[0]):    #2000  
            if data[1] == '04' and data[2] == '14':     #4 14 
                high.append(float(row[-1]))     # high  
                low.append(float(row[-2]))      #  low  

plt.rc('font',family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('    ') # 

plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.legend() #
plt.show()
```
