```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
df
```

```python
sns.stripplot(data=df, x="origin", y="mpg")
```

```python
plt.figure(figsize=(10,4)) #warning massage 삭제시키기
sns.swarmplot(data=df, x="origin", y="mpg",size=3)
```

```python
sns.catplot(data=df, x="origin",y="mpg",col = "cylinders", col_wrap=3) #서브플롯 구성
```

```python
sns.catplot(data=df, x="origin",y="mpg",col = "cylinders", col_wrap=3,kind = "violin")
```

```python
sns.catplot(data=df, x="origin",col = "cylinders", col_wrap=3,kind="count") #countplot은 x,y축중 하나만 있어야함
```

```python
sns.catplot(data=df,kind="box", aspect=3)
```
