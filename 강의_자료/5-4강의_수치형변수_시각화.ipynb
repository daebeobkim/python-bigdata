```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
```

```python
sns.displot(data=df, x="mpg",kde=True,rug=True) #displot을 통해 히스토그램과 kdeplot(곡선)그리기 rud(데이터 빈도확인)
```

```python
sns.kdeplot(data=df, x="mpg",shade = True, cut=5) #shade는 색상
sns.rugplot(data=df, x="mpg")
```

```python
df["mpg"].agg(["skew","kurt"]) #얼마나 치우쳐져있는지 얼마나 뾰족한지
```

```python
sns.boxplot(data=df, x="mpg") #사분위 수 표현하기
```
