```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
```

```python
df["mpg"].unique() #수치형 변수 mpg의 uniqe값 보기(nunique를 사용시 유니크값 수 확인)
```

```python
_ = df.hist() #히스토그램 출력 _사용으로
```

```python
_ = df.hist(figsize = (10,8),bins = 50) #figsize 넓게 그리기 bins 범위개수
```

```python
df.skew() #수치변수에 대한 왜도 구하기 양수값일시 왼쪽에 치우쳐있음 평균과 중강값에 같으면 왜도는 0
```

```python
df.kurt() #첨도 확률 분포의 뽀족한 정도를 나타내는 척도 3보다 작을경우 산포는 정규분포보다 더 뾰족한 분포
```
