```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
sns.scatterplot(data=df,x="weight",y="mpg") #scatterplot을 통해 2개의 수치변수 비교
```

```python
sns.regplot(data=df,x="weight",y="mpg") #regplot으로 회귀선
```

```python
sns.residplot(data=df,x="weight",y="mpg")
```

```python
sns.scatterplot(data=df,x="weight",y="mpg",hue="origin") #scatterplot을 통해 2개의 수치변수 비교하기
```

```python
sns.lmplot(data=df,x="weight",y="mpg",hue="origin",col="origin") #limplot을 통해 범주값에 따라 색상, 서브플롯 그리기
```
