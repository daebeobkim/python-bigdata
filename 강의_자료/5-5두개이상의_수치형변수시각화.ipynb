```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
sns.scatterplot(data=df,x="weight",y="mpg") #scatterplot을 통해 2개의 수치변수 비교
```

```python
sns.regplot(data=df,x="weight",y="mpg") #regplot으로 회귀선
```

```python
sns.residplot(data=df,x="weight",y="mpg")
```

```python
sns.scatterplot(data=df,x="weight",y="mpg",hue="origin") #scatterplot을 통해 2개의 수치변수 비교하기
```

```python
sns.lmplot(data=df,x="weight",y="mpg",hue="origin",col="origin") #limplot을 통해 범주값에 따라 색상, 서브플롯 그리기
```

```python
sns.jointplot(data=df,x="weight",y="mpg",kind="hex") #해당변수의 히스토그램을 같이 그림
```

```python
df_sample = df.sample(100)
df_sample.shape
```

```python
sns.pairplot(data=df_sample,hue="origin") #모든 수치변수를 짝을지어 그려준다 
```

```python
sns.lineplot(data=df,x="model_year",y="mpg",ci=None)
```

```python
sns.relplot(data=df,x="model_year",y="mpg",hue="origin",col="origin") #relplot으로 수치형 변수에 따라 서브플롯 생성
```

```python
sns.relplot(data=df)
```

```python
sns.relplot(data=df,x="model_year",y="mpg",hue="origin",col="origin",kind="line")
```
